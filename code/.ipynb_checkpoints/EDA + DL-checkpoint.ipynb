Data was collected for short period of time (one week), so it misses out seasonality characteristics which I think it would be potential predictor. With the limited data I could only look into weekly features such as days of week, and part of days (morning, evening etc.)

```python
#Function to categorize time into parts of day
def get_part_of_day(x):
    if (x > 4) and (x <= 8):
        return 'Early Morning'
    elif (x > 8) and (x <= 12 ):
        return 'Morning'
    elif (x > 12) and (x <= 16):
        return'Noon'
    elif (x > 16) and (x <= 20) :
        return 'Eve'
    elif (x > 20) and (x <= 24):
        return'Night'
    elif (x <= 4):
        return'Late Night'

tweet['day_of_week'] = tweet['tweet_created'].dt.day_name()                   # 0 = monday
tweet['part_of_day'] = tweet['tweet_created'].dt.hour.apply(get_part_of_day)
```

The text data, primary sources of our sentiment, contains noises such as emoji and hyperlink. Before proceeding to further analysis, I'll clean the data to make it more consistent and understandable by machine. 

```python
def cleaning(text):
    """
    Take a raw tweet as input then clean it into ready-to-use text
    """
    
    # 1. Convert all text to lowercase,
    lower = text.lower()
    
    # 2. Remove hyperlink and special characters except hash sign
    no_html = re.sub(r'(https?:\/\/)?([\da-z\.-]+)\.([a-z\.]{2,6})([\/\w \.-]*)', '', lower)
    letters_only = re.sub("[^#a-zA-Z0-9]", " ", no_html).split()
    
    # 3. Segmenting hashtags
    normal_words = [word for word in letters_only if (word.startswith("#") == False)]
    hashtag_words = [word for word in letters_only if word.startswith("#")]
    hashtag_split = [segment(hashtag) for hashtag in hashtag_words]
    nohash_words = []
    count = 0
    for word in letters_only:
        if word not in hashtag_words:
            nohash_words.append(word)
        else:
            nohash_words.extend(hashtag_split[count])
            count += 1
    
    # 4. Remove mentioning specific airlines
    airline = ['virginamerica', 'united', 'unitedairways', 'southwest', 'southwestair' 'delta', 'usairways',
       'american','americanair','jetblue',]
    useful_words = [word for word in nohash_words if word not in airline]
    
    # 5. remove stopwords
    stops = set(stopwords.words('english'))
    meaningful_words = [word for word in useful_words if word not in stops]
    
    # 6. Lemmatize
    lem = WordNetLemmatizer()
    lem_words = [lem.lemmatize(word) for word in meaningful_words]
    
    return(" ".join(lem_words))
```

## Metadata Modeling

### One Hot Encoding and Standardizing 

For the meta data, I got 2 numerical columns (char_length & word_length) and 2 categorical columns (day_of_week & part_of_day) which need different preprocessing steps as follow

- For the numerical attributes, I'll [standardize](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html) them.
- For the categorical attributes, I'll [one-hot encode](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html)  them

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)
ohe_columns = ['day_of_week', 'part_of_day']
ohe.fit(X_train[ohe_columns])

#Dropping the text columns
X_train_meta = X_train.drop(columns='text')

#One hot encoding the categorical columns
X_train_meta = pd.DataFrame(ohe.transform(X_train[ohe_columns]),
                           columns=ohe.get_feature_names(X_train[ohe_columns].columns), index= X_train.index)
X_train_meta = pd.concat([X_train_meta, X_train.drop(columns = ohe_columns+['text'])], axis=1)

# Standardize numerical columns
ss = StandardScaler()
ss.fit(X_train_meta[['char_length','word_length']])
X_train_meta.loc[:,['char_length','word_length']] = ss.transform(X_train_meta[['char_length','word_length']])

X_train_meta.head(3)
print(X_train_meta.shape)
```

```python
#Dropping the text columns
X_test_meta = X_test.drop(columns='text')

#One hot encoding the categorical columns
X_test_meta = pd.DataFrame(ohe.transform(X_test[ohe_columns]),
                           columns=ohe.get_feature_names(X_test[ohe_columns].columns), index= X_test.index)
X_test_meta = pd.concat([X_test_meta, X_test.drop(columns = ohe_columns+['text'])], axis=1)

# Standardize numerical columns
X_test_meta.loc[:,['char_length','word_length']] = ss.transform(X_test_meta[['char_length','word_length']])

X_test_meta.head(3)
```

For the text data, I'll be looking only to tweet message and following steps will be done to preprocess the text data before feeding into the model

1. [Tokenize the tweet](https://keras.io/api/preprocessing/text/) : vectorize a text corpus, by turning each text into either a sequence of integers (each integer being the index of a token in a dictionary)

1. [Embedding using Standford Glove](https://nlp.stanford.edu/projects/glove/) : to represent each token by the pre-trained word-vector

### Employ WordEmbedding

**Word embedding** is a term used for the representation of words for text analysis, typically in the form of a real-valued vector that encodes the meaning of the word such that the words that are closer in the vector space are expected to be similar in meaning

Next, we compute an index mapping words to known embeddings, by parsing the data dump of pre-trained embeddings:







```python
#Read the pre-trained word-vectors

embeddings_index = {}
GLOVE_PATH = '../glove_twit/'
f = open(os.path.join(GLOVE_PATH, 'glove.twitter.27B.100d.txt'))
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```

```python
# Mapping words to known embedddings weight

embedding_matrix = np.zeros((len(word_index) + 1, 100))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        # words not found in embedding index will be all-zeros.
        embedding_matrix[i] = embedding_vector
```

### Keras Modelling

```python
#Early stopping callback
early_stop = EarlyStopping(monitor = 'val_loss', 
                           min_delta = 0, 
                           patience = 10,          #number of epochs to wait (if val_loss go up) before terminating the training
                           mode = 'auto',          #automatically decide the direction of monitoring metrics (if it's val_loss it should go down, it it's val_acc it should go up)
                           verbose=0,
                           restore_best_weights=True
                          )

#Reduce LR on Plateau
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss', factor=0.1, patience=8, verbose=0,
    mode='auto', min_delta=0.0001, cooldown=0, min_lr=0.00005,
)

def scheduler(epoch, lr):
    if epoch < 5:
        return lr
    else:
        return lr * tf.math.exp(-0.05)
```

```python
# Metadata model architecture
l2_reg = 0.00075

# Input layer, intake the input sequence (all of which have consistent length at 40 tokens).
input_text = Input(shape=(20,))               

# Embedding layer, Mapping each word tokens into 100-dimensional vector space of pre-trained word vectors
# and freeze the weights to prevent further weight update.
embedding_layer = Embedding(len(word_index)+1, 100,
                            weights = [embedding_matrix], trainable = False)(input_text)

# Adding Dropout to regularize the model
sp_dropout_text = SpatialDropout1D(0.20)(embedding_layer)

# Convolutional layer, using different filter to extract features from the text
conv1_text = Conv1D(32, 5, activation = 'relu', kernel_regularizer=regularizers.l2(l2_reg))(sp_dropout_text)

# Max-pooling layer, combining significant features
max_pool_text = MaxPooling1D(pool_size=5)(conv1_text)

# LSTM layer, keeping medium-long term memory
lstm_text = LSTM(32, kernel_regularizer=regularizers.l2(l2_reg),
                 dropout = 0.2, recurrent_dropout = 0.2)(max_pool_text)

# Dense layer
dense_text_1 = Dense(16, activation = 'relu')(lstm_text)

# Output layer, predicting the sentiment class
output_text = Dense(1, activation = 'sigmoid')(dense_text_1)

model_text = Model(inputs=input_text, outputs=output_text)
model_text.compile(loss='binary_crossentropy', optimizer = Adam(learning_rate = 0.0005), metrics=['accuracy'])
model_text.summary()

plot_model(model_text, to_file='model_text.png', show_shapes=True, show_layer_names=True)
```

```python
model_text_history = model_text.fit(X_train_padded,
                   y_train,
                   epochs = 100,
                   batch_size = 512,
                   validation_data = (X_test_padded, y_test),
                   verbose = 1,
                   callbacks = [early_stop, reduce_lr,
                                LearningRateScheduler(scheduler)
                                ]
                                         )
```

```python
# Check out our train loss and validation loss over epochs.
train_loss = model_text_history.history['loss']
test_loss = model_text_history.history['val_loss']

# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_loss, label='Training Loss', color='#185fad')
plt.plot(test_loss, label='Validation Loss', color='orange')

# Set title
plt.title('Training and Validation Loss by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Binary Crossentropy', fontsize = 18)


plt.legend(fontsize = 18);
```

```python
# Check out our train accuracy and validation accuracy over epochs.
train_acc = model_text_history.history['accuracy']
test_acc = model_text_history.history['val_accuracy']

# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_acc, label='Training Accuracy', color='#185fad')
plt.plot(test_acc, label='Validation Accuracy', color='orange')

# Set title
plt.title('Training and Validation Accuracy by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Accuracy', fontsize = 18)

plt.legend(fontsize = 18);
```

## Combining two models

Lastly, as I have combined the machine learning model trained on different subsets of feature. I'll do the same process with our deep learning models.

```python
# Concat layer, concatenating the output from the layer before output of each model and combine them to form new model.
concat_layer = Concatenate()([dense_meta_2, dense_text_1])

# Output layer, predicting the probability of sentiment class
output_combine = Dense(1, activation='sigmoid')(concat_layer)
model_combine = Model(inputs=[input_meta, input_text], outputs=output_combine)
model_combine.compile(loss='binary_crossentropy', 
                      optimizer = Adam(learning_rate = 0.0005), metrics=['accuracy'])

#Model summary
model_combine.summary()
plot_model(model_combine, to_file='model_combine.png', show_shapes=True, show_layer_names=True)
```

```python
model_combine_history = model_combine.fit(x = [X_train_meta,X_train_padded],
                   y = np.array(y_train),
                   epochs = 100,
                   batch_size = 512,
                   validation_data = ([X_test_meta,X_test_padded], y_test),
                   verbose = 1,
                   callbacks = [early_stop, reduce_lr
                                ])
```

```python
# Check out our train loss and validation loss over epochs.
train_loss = model_combine_history.history['loss']
test_loss = model_combine_history.history['val_loss']

# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_loss, label='Training Loss', color='#185fad')
plt.plot(test_loss, label='Validation Loss', color='orange')

# Set title
plt.title('Training and Validation Loss by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Binary Crossentropy', fontsize = 18)


plt.legend(fontsize = 18);
```

```python
# Check out our train accuracy and validation accuracy over epochs.
train_acc = model_combine_history.history['accuracy']
test_acc = model_combine_history.history['val_accuracy']

# Set figure size.
plt.figure(figsize=(12, 8))

# Generate line plot of training, testing loss over epochs.
plt.plot(train_acc, label='Training Accuracy', color='#185fad')
plt.plot(test_acc, label='Validation Accuracy', color='orange')

# Set title
plt.title('Training and Validation Accuracy by Epoch', fontsize = 25)
plt.xlabel('Epoch', fontsize = 18)
plt.ylabel('Accuracy', fontsize = 18)

plt.legend(fontsize = 18);
```

```python
model_combine.evaluate([X_test_meta,X_test_padded], y_test, batch_size=512)
```

With the combining prediction power of metadata model and text model, the model accuracy has been marginally improved from 83.5% (text model) to 84.1%. This might due to our metadata model is not very informative. However it still be useful.
